## Use matplotlib to plot the following equation:

### y = x**2 - x + 2
### Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
%matplotlib inline

xs = range(-100,100)
ys = [(x ** 2 - x + 2) for x in xs]


plt.plot(xs, ys)
plt.annotate('look, an arrow!', xy=(0, 0), xytext=(-40, 4000), fontsize=18,
             arrowprops={'facecolor': 'red'})
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import math

plt.figure(figsize=(10, 12))
plt.suptitle(f'x = {x2}')

x2 = range(2, 15)
# x3 = range(-100, 100)
y1 = [math.sqrt(x) for x in x2]
y2 = [x ** 3 for x in x2]
y3 = [math.tan(x) for x in x2]
y4 = [2 ** x for x in x2]

# plt.subplot(221)
plt.scatter(x2, y1, label='sqrt x', color='red')
plt.plot(x2, y1, color='red')
plt.title('sqrt(x)')

# plt.subplot(222)
plt.scatter(x2, y2, label='x^3', color='green')
plt.plot(x2, y2, color='green')
plt.title('x^3')

# plt.subplot(223)
plt.scatter(x2, y3, label='tan(x)', color='blue')
plt.plot(x2, y3, color='blue')
plt.title('tan(x)')

# plt.subplot(224)
plt.scatter(x2, y4, label='2^x', color='purple')
plt.plot(x2, y4, color='purple')
plt.title('2^x')
plt.legend()
plt.show()
```
